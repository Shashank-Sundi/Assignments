### DATA DESCRIPTION

Features:

1.	months_as_customer: It denotes the number of months for which the customer is associated with the insurance company.

21.	incident_severity: The severity of the incident.
22.	authorities_contacted: Which authority was contacted.
23.	incident_state: The state in which the incident took place.
24.	incident_city: The city in which the incident took place. 
25.	incident_location: The street in which the incident took place.
26.	incident_hour_of_the_day: The time of the day when the incident took place.
27.	property_damage: If any property damage was done.
28.	bodily_injuries: Number of bodily injuries.
29.	Witnesses: Number of witnesses present.
30.	police_report_available: Is the police report available.
31.	total_claim_amount: Total amount claimed by the customer.
32.	injury_claim: Amount claimed for injury
33.	property_claim: Amount claimed for property damage.
34.	vehicle_claim: Amount claimed for vehicle damage.
35.	auto_make: The manufacturer of the vehicle
36.	auto_model: The model of the vehicle. 
37.	auto_year: The year of manufacture of the vehicle.

Whether the claim is fraudulent or not.

```python
# we can observe that-

# Following features should be of type datetime , but are identified as object or numeric
# policy_bind_date , incident_date , auto_year

# We convert them to datetime cuz , We want to create new features from them , instead of dropping them off------Features will be created later on
# claim_duration = policy_bind_date -incident_date ---- then remove the two columns from data , but keep them during html form input
# car_duration = auto_year - incident_date  --- then remove the two columns from data , but keep them during html form input

data.auto_year=pd.to_datetime(data.auto_year,format="%Y")
data.policy_bind_date=pd.to_datetime(data.policy_bind_date)
data.incident_date=pd.to_datetime(data.incident_date)

# Remove following columns-
# _c39 -- contains all nan values , so drop them
# insured_zip, policy_number , incident_location ---- Since all the values are unique and can be unique for every person , 
#                                                    they don't provide any information ; so, we'll drop these features

data=data.drop(['insured_zip', 'policy_number' , 'incident_location', '_c39'],axis=1)
```

### DATA CLEANING & NULL VALUE IMPUTATION

```python
pd.DataFrame(data=data.isnull().sum().values,index=data.isnull().sum().index).T
```

```python
# in above cell, no null values are displayed , cuz the nan values are in the form of an invalid type - ?

data=data.replace('?',np.nan)
pd.DataFrame(data=data.isnull().sum().values,index=data.isnull().sum().index).T
```

```python
# getting columns with null values

null_val_cols=data.isnull().sum()[data.isnull().sum()>0].index
null_val_cols
```

```python
for i in null_val_cols:
    print(data[i].value_counts().head(10))
```

```python
data['collision_type'].isnull().sum()
```

```python
# Performing Random Sample Imputation for the above columns 

for col in null_val_cols:
    rand_samples=data[col].dropna().sample(n=data[col].isnull().sum())
    rand_samples.index=data[col][data[col].isnull()].index
    data.loc[rand_samples.index,col]=rand_samples
```

```python
# All the null values have been imputed 

pd.DataFrame(data=data.isnull().sum().values,index=data.isnull().sum().index).T
```

```python
# Checking Normality of numerical columns

from scipy import stats

fig,ax=plt.subplots(5,3,figsize=(20,15))

i,j=0,0
for col in num_index:
    stats.probplot(data[col],plot=ax[i,j])
    ax[i,j].set_title(col)
    j+=1
    if j==3:
        i+=1
        j=0

fig.tight_layout()

```
